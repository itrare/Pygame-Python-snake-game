```python
import pygame
import time
import random as rd
import numpy as np
score=0
def display_snake(snake_head, snake_position, display):
    pygame.draw.rect(display, (255, 238, 0), pygame.Rect(snake_head[0],snake_head[1],10,10))
    for position in snake_position[1:]:
        pygame.draw.rect(display,(255,0,0),pygame.Rect(position[0],position[1],10,10))
def display_food(food_position, display):
    pygame.draw.rect(display,(0,255,0),pygame.Rect(food_position[0],food_position[1],10,10))
def wall_block(snake_head, snake_position):
    if( snake_head[0]<0 or snake_head[0]>=400 or snake_head in snake_position[1:]):
        return False
    elif( snake_head[1]<20 or snake_head[1]>=420):
        return False
    return True
def food_taken(food_position):
    global score
    score+=1
    food_position = [rd.randrange(1,40)*10, rd.randrange(21,42)*10]
    return food_position, score
def init_values():
    snake_head = [100,100]
    snake_position = [[100,100],[90,100],[80,100]]
    food_position = [rd.randrange(1,40)*10, rd.randrange(21,42)*10]
    return snake_head, snake_position, food_position

def generate_snake(snake_head, snake_position, food_position, button_direction):
    if(button_direction==0):
        snake_head[1]+=10
    elif(button_direction==1):
        snake_head[1]-=10
    elif(button_direction==2):
        snake_head[0]-=10
    elif(button_direction==3):
        snake_head[0]+=10
    else:
        pass
    if(snake_head == food_position):
        food_position, score = food_taken(food_position)
        snake_position.insert(0,list(snake_head))
    else:
        snake_position.insert(0,list(snake_head))
        snake_position.pop()
    
    return snake_head, snake_position, food_position
    
def get_button_direction(event, button_direction):
    save_sirdir = button_direction
    if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT and button_direction != 3:
                button_direction = 2
            if event.key == pygame.K_RIGHT and button_direction != 2:
                button_direction = 3
            if event.key == pygame.K_UP and button_direction != 0:
                button_direction = 1
            if event.key == pygame.K_DOWN and button_direction != 1:
                button_direction = 0
    else:
            button_direction = save_sirdir
        
    return button_direction

def rcrawl_snake( display, button_direction):
    
    CRAWLING = True
    snake_head, snake_position, food_position = init_values()

    while CRAWLING :
        
        CRAWLING = wall_block(snake_head, snake_position)
   
        display.fill((255,255,255))
        text = font.render("Score - "+str(score), True,black,(222, 221, 211))
        display.blit(text, textRect) 
    
        display_snake(snake_head, snake_position, display)
        display_food(food_position, display)
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                CRAWLING = False
            button_direction = get_button_direction(event, button_direction)
        
        snake_head, snake_position, food_position = generate_snake(snake_head, snake_position, food_position, button_direction)
        

        pygame.display.update()
        clock.tick(10)

display_width = 400
display_height = 420

green = (0,255,0)
red = (255,0,0)
black = (0,0,0)
white = (255,255,255)

pygame.init()
display = pygame.display.set_mode((display_width,display_height))
clock = pygame.time.Clock()
display.fill(white)
pygame.display.set_caption('Show Text')
font = pygame.font.Font('freesansbold.ttf', 16)
text = font.render("Score -", True,black,(222, 221, 211))
textRect = text.get_rect()
textRect.center = (200,10)
display.blit(text, textRect)
pygame.display.update()

rcrawl_snake( display, 0)
time.sleep(10)
```
